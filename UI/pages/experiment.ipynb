```python
import streamlit as st
import os
import glob
import pandas as pd
from config.settings import SAVE_PATH, UI_PATH
from UI.pages.schedule_plots import get_plot, last_day
```

```python
# ---- Discover schedule files ----
schedule_files = [os.path.basename(p) for p in glob.glob(os.path.join(SAVE_PATH, "metrics", "20250617*.csv"))]

# ---- Ensure plots are generated once ----@
@st.cache_resource
def generate_plots(files):
    for sched_name in files:
        base_name = os.path.splitext(sched_name)[0]
        plot_path = os.path.join(UI_PATH, f"{base_name}.png")
        dist_path = os.path.join(UI_PATH, f"{base_name}_distribution.png")
        if not os.path.exists(plot_path) or not os.path.exists(dist_path):
            get_plot(sched_name, base_name)
            last_day(sched_name, base_name)

generate_plots(schedule_files)

# ---- Load metrics data ----@
@st.cache_data
def load_schedule_data(files):
    data = []
    for sched_name in files:
        base_name = os.path.splitext(sched_name)[0]
        # Load metrics and trim to first 17 columns
        df = pd.read_csv(os.path.join(SAVE_PATH, "metrics", sched_name))
        df = df[df.columns[:17]]

        # Load schedule to identify slots
        sched = pd.read_csv(os.path.join(SAVE_PATH, 'schedules', sched_name))
        slots = sorted(sched['slot'].unique())
        max_slot = max(slots)
        df['max_slot'] = max_slot

        # Create one-hot indicator columns for slots 1 through max_slot
        for slot_num in range(1, max_slot + 1):
            df[str(slot_num)] = 1 if slot_num in slots else 0

        # Take the first row of metrics and flatten to dict
        m = df.iloc[0]
        data.append({
            "name": base_name,
            "metrics": m.to_dict()
        })

    return data


schedule_data = load_schedule_data(schedule_files)

# ---- Path for likes CSV ----
likes_path = os.path.join(SAVE_PATH, 'like.csv')
if os.path.exists(likes_path):
    likes_df = pd.read_csv(likes_path)
else:
    likes_df = pd.DataFrame(columns=['exam'])

# ---- Extract metric names ----
metric_names = list(schedule_data[0]['metrics'].keys()) if schedule_data else []

# ---- Layout: filters vs results ----
left_col, right_col = st.columns([1.2, 4])

# ---- Filters with dynamic slider ranges ----
filtered_data = schedule_data.copy()
filtered_data
```
